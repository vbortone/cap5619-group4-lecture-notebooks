#### SciPy

```python
from scipy import sparse

# Create a 2D NumPy array with a diagonal of ones, and zeros everywhere else
eye = np.eye(4)
print("NumPy array:\n", eye)

# This line creates a 4x4 identity matrix using the eye function from NumPy. 
# An identity matrix is a square matrix in which all the elements of the 
# principal diagonal are ones, and all other elements are zeros. 
# In this case, you're creating a 4x4 identity matrix.
```

```python
# Convert the NumPy array to a SciPy sparse matrix in CSR format
# Only the nonzero entries are stored
sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy sparse CSR matrix:\n", sparse_matrix)
# This line converts the dense matrix eye into a sparse matrix using the 
# Compressed Sparse Row (CSR) format. The CSR format is a memory-efficient 
# representation of sparse matrices that stores only the non-zero elements and their indices.
```

```python
# Another way to create a sparse matrix in COO (Coordinate) format 
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO representation:\n", eye_coo)

# The line of code you provided is creating a sparse matrix in Coordinate (COO) 
# format using SciPy's coo_matrix function. The COO format represents a sparse matrix 
# by storing the non-zero values along with their row and column indices
```

#### Building Your First Model: k-Nearest Neighbors


`KNeighborsClassifier` is a supervised machine learning algorithm included in the scikit-learn library. It is a type of instance-based or lazy learning algorithm, specifically a k-nearest neighbors (KNN) classifier. The algorithm is used for classification tasks, where the goal is to predict the class or category of a new data point based on its proximity to other known data points in the feature space.



Here's a brief overview of how `KNeighborsClassifier` works:

1. **k-Nearest Neighbors (KNN):**
   - The central idea behind KNN is to classify a new data point by looking at the classes of its k nearest neighbors in the feature space. The "nearest" neighbors are determined based on a distance metric, often Euclidean distance.

2. **Hyperparameter - k:**
   - The parameter "k" represents the number of neighbors to consider when making a prediction. It is a hyperparameter that needs to be specified before training the model.

3. **Training:**
   - During the training phase, the model memorizes the training dataset. No explicit training is involved in the KNN algorithm. The training data itself serves as the "knowledge" for making predictions.

4. **Prediction:**
   - To make a prediction for a new data point, the algorithm calculates the distances between the new point and all points in the training set. It then identifies the k nearest neighbors and assigns the class that is most prevalent among those neighbors to the new point.

5. **Decision Rule:**
   - In the case of ties or equal votes among multiple classes, a decision rule is applied. Common decision rules include simple majority voting or weighted voting based on the distances of neighbors.


In this example, the `KNeighborsClassifier` is instantiated with `n_neighbors=3` (considering 3 nearest neighbors). The model is trained on the training data (`X_train`, `y_train`), and predictions are made on the test set (`X_test`). The accuracy of the model is then evaluated using the true labels (`y_test`).

```python
from sklearn.neighbors import KNeighborsClassifier
# Instantiating the classifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# training the classifier on the train set
knn.fit(X_train, y_train)
```

## Classification Vs Regression

| Aspect                   | Classification                                   | Regression                                       |
|--------------------------|-------------------------------------------------|-------------------------------------------------|
| **Nature of Output**     | Categorical (discrete) labels or classes        | Continuous (real) values                        |
| **Output Interpretation**| Predicts the class or category of an observation| Predicts a numeric value                         |
| **Example**              | Spam or not spam, 0 or 1, dog, cat, or bird      | House price, temperature, stock price, GPA       |
| **Model Type**           | Classification models (e.g., Logistic Regression, Decision Trees) | Regression models (e.g., Linear Regression, Random Forest Regression) |
| **Evaluation Metrics**   | Accuracy, Precision, Recall, F1-score           | Mean Squared Error (MSE), R-squared, Mean Absolute Error (MAE) |
| **Task Objective**       | Assigns data points to predefined categories   | Predicts a continuous value                     |
| **Decision Boundary**    | Separates classes in feature space              | No strict boundary, predicts a value            |
| **Example Libraries**   | Scikit-learn, TensorFlow (for classification tasks) | Scikit-learn, TensorFlow (for regression tasks) |

These are general characteristics, and the specific choice between classification and regression depends on the nature of the problem you are trying to solve and the type of data you have. Classification is used when the output is categorical, and regression is used when the output is continuous.

In FinTech, the choice between classification and regression depends on the specific nature of the problem you are trying to solve and the type of output or prediction you need. Here are some guidelines for when to use classification and when to use regression in FinTech:

**Use Classification When:**

1. **Categorical Outputs:**
   - When the problem involves predicting categories or labels. For example, classifying transactions as fraudulent or non-fraudulent, customers as high-risk or low-risk, or companies as buy/sell/hold.

2. **Default Prediction:**
   - In credit risk assessment, classification models can be used to predict whether a customer is likely to default on a loan (binary classification).

3. **Fraud Detection:**
   - Identifying fraudulent activities, such as credit card fraud or account takeover. The output is typically binary - fraud or non-fraud.

4. **Customer Segmentation:**
   - Segmenting customers into different categories based on their financial behavior, risk profiles, or preferences.

**Use Regression When:**

1. **Numeric Outputs:**
   - When the problem involves predicting a numeric or continuous value. For example, predicting stock prices, interest rates, or housing prices.

2. **Risk Assessment:**
   - In financial risk management, regression can be used to model the expected loss or potential impact of different risk factors.

3. **Valuation Models:**
   - In asset pricing, regression can be employed to build valuation models for various financial instruments.

4. **Time Series Forecasting:**
   - When predicting financial time series data, such as stock prices or currency exchange rates, regression models can be effective.

5. **Portfolio Optimization:**
   - In portfolio management, regression models can help optimize portfolios by predicting the expected returns of different assets.

6. **Credit Scoring:**
   - While classification can be used for credit scoring, regression models can provide a more nuanced approach by predicting a credit score as a continuous variable.

It's important to carefully analyze the problem at hand, the nature of the data, and the goals of the prediction. Sometimes, a combination of both classification and regression techniques may be used in a FinTech application, depending on the complexity of the problem and the desired level of granularity in predictions.

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# instantiate the model and set the number of neighbors to consider to 3
reg = KNeighborsRegressor(n_neighbors=3)
# fit the model using the training data and training targets
reg.fit(X_train, y_train)
```

```python
print("Test set predictions:\n", reg.predict(X_test))
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
# To compute the R^2 regression score on a test set in Python, you can use the score 
# method provided by many regression models in scikit-learn. The score, also known as 
# the coefficient of determination, measures the proportion of the response variable's 
# variance that is captured by the model. A score of 1 indicates a perfect prediction, 
# while a score of 0 indicates that the model does no better than predicting the mean 
# of the target variable.
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    # Train the model on the training set 
    reg.fit(X_train, y_train)
    # predict the output from the trained model
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train),
            reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",
                "Test data/target"], loc="best")
```

```python
from sklearn.linear_model import LinearRegression
# Create the dataset with 60 samples
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
# train the liner regression model on the data
lr = LinearRegression().fit(X_train, y_train)

print("lr.coef_:", lr.coef_)
print("lr.intercept_:", lr.intercept_)
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lr.score(X_test, y_test)))
```

#### Ridge regression

```python
from sklearn.linear_model import Ridge
# Linear least squares with l2 regularization.
ridge = Ridge().fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test, y_test)))
```

```python
# "alpha" controls the regularization strength
# Regularization improves the conditioning of the problem and reduces the variance of the estimates. 
# Larger values specify stronger regularization
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
plt.plot(ridge.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.1")

plt.plot(lr.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-25, 25)
plt.legend()
```

#### Lasso

```python
from sklearn.linear_model import Lasso
# Lasso - Linear Model trained with L1 prior as regularizer
lasso = Lasso().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso.coef_ != 0))
```

```python
# we increase the default setting of "max_iter",
# otherwise the model would warn us that we should increase max_iter.
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso001.coef_ != 0))
```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso00001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso00001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso00001.coef_ != 0))
```

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01")
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha=0.0001")

plt.plot(ridge01.coef_, 'o', label="Ridge alpha=0.1")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)
logreg = LogisticRegression().fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
# C - Inverse of regularization strength; must be a positive float. 
# Like in support vector machines, smaller values specify stronger regularization.
logreg100 = LogisticRegression(C=100).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test, y_test)))
```

```python
# Regression is different from classification, where we predict a value rather the a discrete class label
logreg001 = LogisticRegression(C=0.01).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg001.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg001.score(X_test, y_test)))
```

```python
plt.plot(logreg.coef_.T, 'o', label="C=1")
plt.plot(logreg100.coef_.T, '^', label="C=100")
plt.plot(logreg001.coef_.T, 'v', label="C=0.001")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.legend()
```

```python
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(C=C, solver='liblinear', penalty="l1").fit(X_train, y_train)
    print("Training accuracy of l1 logreg with C={:.3f}: {:.2f}".format(
          C, lr_l1.score(X_train, y_train)))
    print("Test accuracy of l1 logreg with C={:.3f}: {:.2f}".format(
          C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C))

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")

plt.ylim(-5, 5)
plt.legend(loc=3)
```

#### Key differences between Linear Regression, Logistic Regression, Ridge Regression, and Lasso Regression:

| Aspect                   | Linear Regression                                | Logistic Regression                              | Ridge Regression                                | Lasso Regression                                |
|--------------------------|--------------------------------------------------|---------------------------------------------------|--------------------------------------------------|--------------------------------------------------|
| **Type**                 | Regression                                       | Classification                                    | Regression                                      | Regression                                      |
| **Output**               | Continuous values (real numbers)                | Probability of belonging to a class (0 to 1)      | Continuous values (real numbers)                | Continuous values (real numbers)                |
| **Objective Function**   | Minimize Mean Squared Error (MSE)               | Maximize Likelihood Function (cross-entropy)     | Minimize MSE + Regularization Term (L2 norm)    | Minimize MSE + Regularization Term (L1 norm)    |
| **Equation**             | \(y = \beta_0 + \beta_1 x_1 + \ldots + \beta_n x_n\) | \(P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \ldots + \beta_n x_n)}}\) | Similar to linear regression, but with penalty term | Similar to linear regression, but with penalty term |
| **Target Variable**      | Continuous (numeric)                             | Binary or categorical                            | Continuous (numeric)                             | Continuous (numeric)                             |
| **Cost Function**        | Mean Squared Error (MSE)                         | Cross-Entropy Loss                               | MSE + L2 regularization term                   | MSE + L1 regularization term                   |
| **Regularization Term**  | None                                             | None                                              | L2 (Euclidean norm of coefficients)            | L1 (Manhattan norm of coefficients)            |
| **Use Case**             | Predicting a numeric outcome                     | Binary or multiclass classification              | Dealing with multicollinearity in linear regression | Feature selection, handling sparse data         |
| **Interpretability**     | Coefficients represent the change in the dependent variable for a unit change in the independent variable | Coefficients represent the change in the log-odds of the dependent variable for a unit change in the independent variable | Coefficients represent the change in the dependent variable for a unit change in the independent variable with regularization | Coefficients represent the change in the dependent variable for a unit change in the independent variable with regularization |

These are general characteristics, and the choice between linear, logistic, ridge, or lasso regression depends on the nature of the problem, the type of data, and the specific goals of the analysis.
